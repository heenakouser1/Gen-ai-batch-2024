```python
# diamond problem --- ambiguity
# this can be solved by using the method resolution order (MRO)


class Father:
  def propertyofparent(self):
    print("fathers property")
class Mother:
  def propertyofparent(Self):
    print("moms property")
class Son(Father,Mother):
  def propertyofson(Self):
    print("sons property")
```

```python
S = Son()
```

```python
S.propertyofparent()   # even if method are same in both parent, it picks up the father becz that is written first during child class declaration
```

```python
def my_decorator(func):
  def wrapper():
    print("before compuation")
    func()
    print("after compuation")
  return wrapper
```

```python
@my_decorator
def say_hello():
  print("hello")
```

```python
say_hello()
```

```python
say_bye()
```

```python
import time
```

```python
def cal_time(func):
  def wrapper():
    start = time.time()
    func()
    end = time.time()
    print("time taken ", end-start)
  return wrapper
```

```python
@cal_time
def caltime():
  print(100000000000000*3)
```

```python
caltime()
```

```python
# class decorators
class my_decorator:
  def __init__(self,func):
    self.func = func
  def __call__(self):
    print("before compuation")
    self.func()
    print("after compuation")
```

```python
# custom exception

def validatesalary(salary):
  if salary <= 0:
    raise ValidateSalary("Salary is not valid")
  elif salary>=300000000000:
    raise ValidateSalary("Salary is too high")
  else:
    raise ValidateSalary("Salary is valid")
```

```python
class ValidateSalary(Exception):
  def __init__self(self,msg):
    self.msg = msg
```

```python
try:
  salary = int(input("enter the salary"))
  validatesalary(salary)
except ValidateSalary as e:
  print(e)
```

```python
import logging
logging.basicConfig(filename = 'ex1.log',level=logging.INFO, format = '%(asctime)s %(message)s %(levelname)s')
def separate(l):
  updatedlist = []
  updatedstr = []
  for i in l:
    logging.info(" this is info for all i fir int ",i)
    if type(i) == int:
      updatedlist.append(i)
    elif type(i) == list:
      for j in i:
        logging.info(" this is info for all j for list ",j)
        if type(j) == int:
          updatedlist.append(j)
        elif type(j) == str:
          updatedstr.append(j)
    elif type(i) == str:
      logging.info(" this is info for all i fir str ",i)
      updatedstr.append(i)
  return "updated list :" ,updatedlist, "updated str :",updatedstr
```

```python
separate(l)
```
