
Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each

```python
import pandas as pd
import numpy as np

# 2D array can be created by usinh full function from numpy library
m1 = np.full((3,3),7)
print(m1, 'Dimen of m1: ',np.ndim(m1))
```

```python
# 2D array can be created by using ones function from numpy library
m2 = np.ones((3,3),dtype=int) *7
print(m2, 'Dimen of m2: ',np.ndim(m2))
```

```python
# 2D array can be created by using tile function from numpy library
m3 = np.tile(7, (3,3))
print(m3, 'Dimen of m3: ',np.ndim(m3))
```

## np.array, np.asarray and np.asanyarray

```python
# np.array, np.asarray and np.asanyarray

# np.array---- always creates a copy
x = np.array([1,2,3])
n1 = np.array(x)
print(x is n1)
```

```python
# np.asarray -- avoids copying if input is already a compatiable array
y = np.array([1,2,3])
n2 = np.asarray(y)
print(y is n2)

m = np.matrix([1,2,3])
n5 = np.asarray(m)
print(m is n5)
```

```python
# np.asanyarray -- is the most flexible and works with subclasses of ndarray without necessarily creating a copy.
z = np.array([1,2,3])
n3 = np.asanyarray(z)
print( z is n3)

m = np.matrix([1,2,3])
n4 = np.asanyarray(m)
print(m is n4)
```

## Deep copy and shallow copy

```python
# Shallow copy
n1 = np.array([1,2,3])
n2 = n1
n1[0] = 10
print('n1', n1)
print('n2' , n2)
print(n1 is n2)
```

```python
# Deep copy
n1 = np.array([1,2,3])
n2 = n1.copy()
n1[0] = 10
print('n1', n1)
print('n2' , n2)
print(n1 is n2)
```

17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
np.random.seed(10)
x = np.random.randint(1,10,size = 10)
y = np.random.randint(1,10,size = 10)
data = pd.DataFrame({'x':x,'y':y})
data['Quadrant'] = 0

data.loc[(data['x'] > 0.5) & (data['y'] > 0.5), 'Quadrant'] = 1
data.loc[(data['x'] < 0.5) & (data['y'] > 0.5), 'Quadrant'] = 2
data.loc[(data['x'] > 0.5) & (data['y'] < 0.5), 'Quadrant'] = 3
data.loc[(data['x'] > 0.5) & (data['y'] > 0.5), 'Quadrant'] = 4

sns.scatterplot(x = 'x', y = 'y', data=data,color='r',hue = 'Quadrant',palette = 'rainbow')

plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'

```python
import bokeh.io
import bokeh.plotting

from bokeh.plotting import figure
from bokeh.io import output_notebook, show


n1 = np.linspace(1,10,10)
n2 = np.sin(n1)

p = figure(title = "Sine wave function")
p.line(n1,n2)
p.grid.visible = True
p.xaxis.axis_label = 'x-axis'
p.yaxis.axis_label = 'y-axis'
output_notebook()
show(p)
```

19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'

```python
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool

x = ['a','b','c','d']
n1 = np.random.choice(x,size = 10)
n2 = np.random.randint(1,10,size= 10)
df = pd.DataFrame({'H':n1,'K':n2})

p = figure(title = 'Random Categorical Bar Chart',x_axis_label='value choice',
           y_axis_label='values',
           x_range=df['H'].unique())

p.vbar(x= 'H',top= 'K',source = df)
hover = HoverTool(tooltips=[('H', '@H'), ('K', '@K')])
p.add_tools(hover)

output_notebook()
show(p)

# open the notebook in google collab
```
