1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

```python
import plotly.graph_objects as go
import plotly.express as px


np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df['X'], y = df['Y'], z = df['Z'], mode = 'markers' ))
fig.update_layout(
     autosize=False,
     width=800,
     height=800
 )
fig.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

z_re = df['Z'].values.reshape(x.shape)
```

```python
fig = go.Figure()
fig.add_trace(go.Surface( x = df['X'] , y = df['Y'], z = z_re))
fig.update_layout(
     autosize=False,
     width=500,
     height=300
 )
fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(1, 15, 100)
y = np.sin(x)

p = figure(width=600, height=400, title = " sine wave ")
p.xaxis.axis_label = 'x axis'
p.yaxis.axis_label = 'y axis'
p.line(x , y)
output_notebook()
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given dat

```python
from bokeh.models import Quad
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

fig = figure(width=600, height=400, title = 'Histogram')
fig.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
         fill_color="navy", line_color="white", alpha=0.5)
output_notebook()
show(fig)
```

5. Create a Bokeh heatmap using the provided datase

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```

```python
import matplotlib as mpl
normalized_data = (data_heatmap - data_heatmap.min()) / (data_heatmap.max() - data_heatmap.min())
cmap = mpl.cm.get_cmap('rainbow')
colors = [mpl.colors.rgb2hex(cmap(value)) for value in normalized_data.flatten()]


fig = figure(width=600, height=400, title = " heat map")
fig.rect(xx.flatten(), yy.flatten(),color=colors, width=0.1, height=0.1)

output_notebook()
show(fig)
```
